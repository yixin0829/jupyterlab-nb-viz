```python
# read the dataset
df = pd.read_csv('../datasets/2-ds-job-salaries.csv', index_col=0)
```

```python
# split categorical and numerical columns
categorical_columns = ['experience_level', 'employment_type', 'job_title', 'salary_currency', 'employee_residence', 'company_location', 'company_size' ]
categorical = df.select_dtypes(include= "object")

numerical_columns = ['work_year', 'remote_ratio']
numerical = df.select_dtypes(exclude= "object")
```

```python
# linear regression with statsmodels
df_dummies = pd.get_dummies(df[categorical_columns], drop_first=True)
X = pd.concat([numerical,df_dummies], axis = 1)
X = X.dropna(axis=1)
y = df["salary_in_usd"]
import statsmodels.api as sm
X = sm.add_constant(X)
X.reset_index(drop=True, inplace=True)
y.reset_index(drop=True, inplace=True)
y = list(y)
model = sm.OLS(y, X).fit()
print(model.summary())
```
