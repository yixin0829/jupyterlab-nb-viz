```python
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('../datasets/1-ibm-hr.csv')
```

```python
# Attrition
sns.catplot(y="Attrition", data=df, kind="count", palette= "prism")
```

```python
# Attrition and age
sns.swarmplot(x='Attrition', y='Age', data=df)
```

```python
sns.swarmplot(x='Attrition', y='TotalWorkingYears', data=df)
```

```python
sns.boxplot(x="Attrition", y="MonthlyIncome", data=df, palette='Spectral')
```

```python
sns.catplot(x="JobLevel", y="MonthlyIncome", col="JobRole", kind="box", data=df, palette="magma", ci=None)
```

```python
sns.boxplot(x="JobSatisfaction", y="MonthlyIncome", data=df, palette='Spectral')
```

It seems that the income levels of employees with different jobsatisfaction are pretty similar.
The slight difference is that the higher the JobSatifaction, the lower the difference in MonthlyIncome.
