```python
import pandas as pd
import seaborn as sns
import plotly.express as px
df = pd.read_csv('datasets/IBM_HR.csv')
```

```python
# Attrition
sns.catplot(y="Attrition", data=df, kind="count", palette= "prism")
```

```python
# Attrition and age
sns.swarmplot(x='Attrition', y='Age', data=df)
```

```python
sns.swarmplot(x='Attrition', y='YearsWithCurrManager', data=df)
```

```python
sns.boxplot(x="Attrition", y="MonthlyIncome", data=df, palette='Spectral')
```
