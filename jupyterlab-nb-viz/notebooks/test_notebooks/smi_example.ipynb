```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('../datasets/1-ibm-hr.csv')
```

```python
df.head()
```

```python
import plotly.express as px
fig = px.box(df, x="JobLevel", y="MonthlyIncome", color="Attrition", width=600, height=300, category_orders={"Attrition": ["No", "Yes"]})
fig.update_layout(font=dict(size=20))
fig.show()
```

From this boxplot, we can see how every job level almost has a corresponding range of monthly income.

The difference in median monthly income (No Attrition - Attrition ) is greatest for Job Level= 4 followed by Job Level= 1.

A possible reason for high attrition rate at Job Level 1 may be that they feel it is unfair to be receiving a lower monthly income than others in the same job level.

Also, if Job Level 4 has such a huge difference, why is the attrition rate low?

```python
joblevel4= df.loc[df["JobLevel"]==4]
fig = px.box(joblevel4, x="StockOptionLevel", y="MonthlyIncome", points='all', color="Attrition", title= "JobLevel=4", width=600, category_orders={"Attrition": ["No", "Yes"]})
fig.update_layout(font=dict(size=20))
fig.show()
```

Possible reason for low attrition in Job Level 4:

Even though there is a huge gap in median monthly income, most of the employees there have a high monthly income that may compensate for a lower stock option level. The red, box for those with attrition have no stocks and  a much lower monthly income compared to the rest.
